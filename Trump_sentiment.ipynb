```python
data = pd.read_csv("Donald-Tweets!.csv")
```

```python
tweets = data['Tweet_Text']
```

```python
polarity = []
for i in tweets:
    txt = TextBlob(i)
    polarity.append((txt.sentiment.polarity)*10)
```

```python
columns = ['Tweet','Polarity','Hashtags','twt_favourites_IS_THIS_LIKE_QUESTION_MARK','Retweets']
data = pd.DataFrame(data, columns=columns)
data.head()
```

```python
data['Tweet'] =tweets
data['Polarity'] =pd.DataFrame(polarity)
```

```python
data_by_polarity =data.sort_values(by='Polarity',ascending=False)
```
